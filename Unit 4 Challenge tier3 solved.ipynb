**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
properties = properties.loc[:, ~properties.columns.str.contains('^Unnamed')]
print(properties.shape)

```

```python
tidy = properties.melt(id_vars= 'Date', var_name= 'Area', value_name= 'AveragePrice')
tidy= tidy.dropna(subset= ['AveragePrice'])
tidy.head()
```

```python
#drop rows with NaT in 'Date' column
tidy.dropna(subset=["Date"])
#Ensure AveragePrice is numeric
tidy['AveragePrice'] = pd.to_numeric(tidy['AveragePrice'], errors='coerce')
tidy.head()
```

```python
tidy = tidy.dropna(subset=["Date", "AveragePrice"]).reset_index(drop=True)
tidy.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio (London_borough):
    #select rows where Area is London_borough and Year is 1998 or 2018
    df = tidy[(tidy['Area'] == London_borough) & (tidy['Year'].isin([1998, 2018]))]
    #get the average price for that borough in 1998 and 2018
    avg_prices = df.groupby('Year')['AveragePrice'].mean()
    #calculate the price ratio
    price_ratio = avg_prices.loc[2018] / avg_prices.loc[1998]
    return price_ratio

#iterate through all unique boroughs in Area column and create a dictionary with borough as key and price ratio as value
price_ratios = {}   
for borough in tidy['Area'].unique():
    try:
        ratio = create_price_ratio(borough)
        price_ratios[borough] = ratio
    except KeyError:
        #skip boroughs that don't have data for both years
        continue
#convert dictionary to dataframe
price_ratios_df = pd.DataFrame(list(price_ratios.items()), columns=['Borough', 'PriceRatio'])
#sort dataframe by PriceRatio in descending order
price_ratios_df = price_ratios_df.sort_values(by='PriceRatio', ascending=False)
price_ratios_df.head(10)


```

```python
price_ratios_df.tail(10)
```

```python
# Plot the top 15 boroughs
top_15 = price_ratios_df.head(15)
plt.figure(figsize=(10, 8))
plt.barh(top_15['Borough'], top_15['PriceRatio'])
plt.xlabel('Price Ratio')
plt.title('Top 15 London Boroughs by House Price Ratio (2018 vs 1998)')
plt.gca().invert_yaxis() # To display the highest ratio at the top
plt.show()
```
